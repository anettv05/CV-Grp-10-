```python
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Input, Dense, GlobalAveragePooling2D, Dropout, 
                                     BatchNormalization, Activation, Lambda, 
                                     RandomFlip, RandomRotation, RandomZoom)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
import matplotlib.pyplot as plt

# =============================
# 1. Distributed Training & Dataset Setup
# =============================

# Set up distributed training strategy
strategy = tf.distribute.MirroredStrategy()
print(f"Number of devices: {strategy.num_replicas_in_sync}")

# Set dataset directory and image parameters
dataset_dir = "/kaggle/working/PlantVillage-Dataset/raw/color"
IMG_SIZE = (224, 224)
BATCH_SIZE = 32 * strategy.num_replicas_in_sync  

# Create training and validation datasets (80/20 split)
train_ds = tf.keras.utils.image_dataset_from_directory(
    dataset_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    validation_split=0.2,
    subset="training",
    seed=42,
    label_mode='categorical'
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    dataset_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    validation_split=0.2,
    subset="validation",
    seed=42,
    label_mode='categorical'
)

# Disable auto-sharding to keep batches consistent
options = tf.data.Options()
options.experimental_distribute.auto_shard_policy = tf.data.experimental.AutoShardPolicy.OFF
train_ds = train_ds.with_options(options)
val_ds = val_ds.with_options(options)

# Data augmentation: random flip, rotation, and zoom
data_augmentation = tf.keras.Sequential([
    RandomFlip("horizontal"),
    RandomRotation(0.2),
    RandomZoom(0.1)
])

# Apply augmentation only on the training data and prefetch for speed
train_ds = train_ds.map(lambda x, y: (data_augmentation(x, training=True), y),
                        num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.prefetch(tf.data.AUTOTUNE)
val_ds = val_ds.prefetch(tf.data.AUTOTUNE)

# Debug: Inspect dataset shapes
for images, labels in train_ds.take(1):
    print("Train batch shapes:", images.shape, labels.shape)
for images, labels in val_ds.take(1):
    print("Validation batch shapes:", images.shape, labels.shape)

# =============================
# 2. Build the InceptionV3-Based Model & Phase 1 Training (Head Training)
# =============================

with strategy.scope():
    # Define model input
    inputs = Input(shape=(224, 224, 3))
    
    # Preprocess inputs for InceptionV3: scales pixel values from [0, 255] to [-1, 1]
    x = Lambda(tf.keras.applications.inception_v3.preprocess_input)(inputs)
    
    # Create the InceptionV3 base model with ImageNet weights (exclude the top classifier)
    base_model = InceptionV3(weights='imagenet', include_top=False, input_tensor=x)
    
    # Phase 1: Freeze the entire base model
    base_model.trainable = False
    
    # Build the classification head
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(512)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(0.5)(x)
    x = Dense(256)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(38, activation='softmax')(x)  # 38 classes

    model = Model(inputs=inputs, outputs=outputs)
    
    # Compile the model for Phase 1 with a moderately high learning rate
    model.compile(optimizer=Adam(learning_rate=1e-4), 
                  loss='categorical_crossentropy', 
                  metrics=['accuracy'])
    
    # Print model summary
    model.summary()

# Callbacks common to both phases
common_callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ModelCheckpoint("best_model_phase1.keras", monitor='val_accuracy', save_best_only=True, mode='max'),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)
]

print("\n--- Phase 1: Training Classification Head ---")
history_phase1 = model.fit(
    train_ds,
    epochs=10,
    validation_data=val_ds,
    callbacks=common_callbacks
)

# =============================
# 3. Phase 2: Fine-Tuning by Unfreezing Late Layers
# =============================

with strategy.scope():
    # Unfreeze later layers (e.g., layers with names containing 'mixed7', 'mixed8', 'mixed9')
    for layer in base_model.layers:
        if any(key in layer.name for key in ['mixed7', 'mixed8', 'mixed9']):
            layer.trainable = True
        else:
            layer.trainable = False
    
    # Re-compile the model with a lower learning rate for fine-tuning
    model.compile(optimizer=Adam(learning_rate=1e-5), 
                  loss='categorical_crossentropy', 
                  metrics=['accuracy'])
    
    # Optional: Print updated model summary to verify trainable layers
    print("\n--- Fine-Tuning: Updated Model Summary ---")
    model.summary()

phase2_callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ModelCheckpoint("best_model_phase2.keras", monitor='val_accuracy', save_best_only=True, mode='max'),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)
]

print("\n--- Phase 2: Fine-Tuning ---")
history_phase2 = model.fit(
    train_ds,
    epochs=20,
    validation_data=val_ds,
    callbacks=phase2_callbacks
)

# =============================
# 4. Evaluate, Save, and Plot Training History
# =============================

# Evaluate the final model on the validation set
test_loss, test_acc = model.evaluate(val_ds)
print(f"\nFinal Test Accuracy: {test_acc:.4f}")

# Save the final fine-tuned model
model.save("plant_disease_inception_finetuned.h5")

# Combine history for plotting (Phase 1 + Phase 2)
def combine_history(hist1, hist2):
    history = {}
    for key in hist1.history.keys():
        history[key] = hist1.history[key] + hist2.history[key]
    return history

combined_history = combine_history(history_phase1, history_phase2)

# Plot accuracy and loss curves
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(combined_history['accuracy'], label='Train Accuracy')
plt.plot(combined_history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title("Accuracy")
plt.subplot(1, 2, 2)
plt.plot(combined_history['loss'], label='Train Loss')
plt.plot(combined_history['val_loss'], label='Validation Loss')
plt.legend()
plt.title("Loss")
plt.show()
```
